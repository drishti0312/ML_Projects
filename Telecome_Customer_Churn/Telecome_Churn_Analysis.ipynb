## EDA

```python
sns.countplot(df,x='gender',palette='Set2')
# The dataset contains an almost equal distribution of male and female customers, indicating gender imbalance is not present in the data.
```

```python
sns.countplot(df,x='SeniorCitizen',palette='Set1')
```

```python
df['SeniorCitizen'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
senior_citizen = df[df['SeniorCitizen'] == 1]
pec_citizen = (len(senior_citizen)/len(df))*100
print("% of senior citizen",pec_citizen)

not_senior_citizen = df[df['SeniorCitizen'] == 0]
pec_not_citizen = (len(not_senior_citizen)/len(df))*100
print("% of not senior citizen",pec_not_citizen)

# Only around 16% customers are senior citizens, showing that the customer base is primarily composed of younger or middle-aged customers.
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()

# Number of churn = Yes is around 1800
```

```python
df.columns
```

```python
sns.histplot(df['tenure'],bins=30,kde=True)
plt.title('Distribution of Customer Tenure')
plt.xlabel("Tenure (Months)")
plt.ylabel("Count")
plt.show()

# Tenure distribution is slightly left-skewed, indicating many customers are relatively new.
```

```python
female = df[df['gender'] == 'Female']
pec = (len(female)/len(df))*100
print(pec)

male = df[df['gender'] != 'Female']
pec_m = (len(male)/len(df))*100
print(pec_m)

# There are equal number of Males and Females 
```

```python
new_customers = df[df['tenure'] <= 12]
percentage = (len(new_customers) / len(df)) * 100
print(percentage)
```
